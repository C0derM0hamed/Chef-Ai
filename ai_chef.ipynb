```python
from langchain.chat_models import init_chat_model
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
llm = init_chat_model(
    "gpt-4o-mini",
    temperature=0.7
)
```

```python
system_prompt = """
You are a chef.

Give meals based on ingredients.
Return JSON only. No markdown..

Format:
[
  {
    "name": "",
    "cooking_time": 0,
    "servings": 0,
    "ingredients": [
      {"name": "", "status": "available or missing"}
    ],
    "instructions": []
  }
]
"""
```

```python
import base64

img = base64.b64encode(open("image.png", "rb").read()).decode()               
```

```python
from langchain_core.messages import HumanMessage
```

```python
message = HumanMessage(content=[
    {"type": "text", "text": system_prompt + " Take ingredients from image "},
    {
        "type": "image_url",
        "image_url": {"url": f"data:image/png;base64,{img}"}
    }
])
```

```python
full_response = ""

for chunk in llm.stream([message]):
    if chunk.content:
        print(chunk.content, end="", flush=True)
        full_response += chunk.content
```

```python
print(full_response)
```

```python
import json

data = json.loads(full_response)
```

```python
for meal in data:
    print(meal["name"])
    print(meal["cooking_time"])
    
    for step in meal["instructions"]:
        print("-", step)
```
